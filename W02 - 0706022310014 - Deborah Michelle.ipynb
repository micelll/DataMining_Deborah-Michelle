```python
# menghapus column yang tidak diperlukan. 
# URL dan title tidak diperlukan untuk analisis
# address hanya gabungan dari district dan city
# property_type karena semua adalah 'rumah'
# building age sama saja dengan year_built. jadi, salah satu dihapus. Namun, year_built terpilih karena building age datanya tidak diperbarui berdasarkan tahun sekarang
df = df.drop(columns=['url', 'title', 'address', 'property_type', 'building_age'])
df.head()
```

```python
# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## Missing Value Handling

```python
# drop rows dimana ads_id null. Alasan didrop adalah karena hanya ada 4 row yang null
df = df.dropna(subset=['ads_id'])
```

```python
# Missing values dari list column di bawah ini diisi dengan 'Unknown' karena bersifat kategorikal
missing = [
    'certificate', 
    'property_condition', 
    'building_orientation', 
    'furnishing'
]

df[missing] = df[missing].fillna('Unknown')
df[missing].isna().sum()
```

```python
# diisi dengan mode karena kolom ini bertipe kategorikal diskrit (jumlah kamar/ lantai berupa bilangan kecil)
mode_cols = ['bedrooms', 'bathrooms', 'floors']
for col in mode_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

# diisi dengan median karena kolom ini numerik dan rentangnya sangat lebar dengan banyak outlier
median_cols = ['land_size_m2', 'building_size_m2', 'year_built']
for col in median_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python
# konversi ke int karena tidak ada nilai desimal
float_cols = [
    'bedrooms', 
    'bathrooms', 
    'land_size_m2', 
    'building_size_m2', 
    'carports', 
    'maid_bedrooms', 
    'maid_bathrooms', 
    'floors',  
    'garages'
]

# fill NaN with 0 then convert to int
df[float_cols] = df[float_cols].fillna(0).astype(int)
```

```python
# memastikan sudah tidak ada missing value
df.isnull().sum()
```

## Detect Outliers

```python
# Pembentukan boxplot untuk mendeteksi outlier pada column numerik
# Berikut adalah column numerik yang memliki outlier yang cukup ekstrim

# Boxplot for 'Jumlah Kamar'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bedrooms'])
plt.title('Jumlah Kamar')
plt.xlabel('Jumlah Kamar')
plt.show()

# Boxplot for 'Jumlah Kamar Mandi'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bathrooms'])
plt.title('Jumlah Km Mandi')
plt.xlabel('Jumlah Km Mandi')
plt.show()

# Boxplot for 'Luas Tanah'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['land_size_m2'])
plt.title('Luas Tanah (m2)')
plt.xlabel('Luas Tanah (m2)')
plt.show()

# Boxplot for 'Luas Bangunan'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['building_size_m2'])
plt.title('Luas Bangunan (m2)')
plt.xlabel('Luas Bangunan (m2)')
plt.show()

# Boxplot for 'Garages'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['garages'])
plt.title('Jumlah Garages')
plt.xlabel('Garages')
plt.show()

# Boxplot for 'Carports'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['carports'])
plt.title('Jumlah Carports')
plt.xlabel('Carports')
plt.show()

# Boxplot for 'Price'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price_in_rp'])
plt.title('Price (in Rp)')
plt.xlabel('Price')
plt.show()
```

```python
# Menghapus outlier berdasarkan visualisasi boxplot
# Bedrooms dan Bathrooms dibatasi 10 saja karena jumlah kamar yang masuk akal pada suatu rumah biasanya tidak lebih dari 10
df = df[df['bedrooms'] <= 10]
df = df[df['bathrooms'] <= 10]

# Banyak luas tanah dan bangunan yang berada jauh dari batas di bawah ini, dan menurut saya
# 1500 dan 2000 adalah batas yang masuk akal untuk menjaga analisis tetap representatif
# untuk mayoritas properti tanpa bias dari propoerti yang sangat besar
df = df[df['land_size_m2'] <= 1500]
df = df[df['building_size_m2'] <= 2000]

# Dibatasi hanya 5 karena saya asumsikan jumlah ini adalah jumlah yang masuk akal untuk suatu rumah
df = df[df['garages'] <= 5]
df = df[df['carports'] <= 5]

# Batas harga ditingkatkan menjadi 70 milyar untuk mengurangi outlier yang sangat ekstrim
df = df[df['price_in_rp'] <= 70000000000]

# Menampilkan kembali boxplot setelah penghapusan outlier
# Boxplot for 'Jumlah Kamar'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bedrooms'])
plt.title('Jumlah Kamar')
plt.xlabel('Jumlah Kamar')
plt.show()

# Boxplot for 'Jumlah Kamar Mandi'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bathrooms'])
plt.title('Jumlah Km Mandi')
plt.xlabel('Jumlah Km Mandi')
plt.show()

# Boxplot for 'Luas Tanah'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['land_size_m2'])
plt.title('Luas Tanah (m2)')
plt.xlabel('Luas Tanah (m2)')
plt.show()

# Boxplot for 'Luas Bangunan'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['building_size_m2'])
plt.title('Luas Bangunan (m2)')
plt.xlabel('Luas Bangunan (m2)')
plt.show()

# Boxplot for 'Garages'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['garages'])
plt.title('Jumlah Garages')
plt.xlabel('Garages')
plt.show()

# Boxplot for 'Carports'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['carports'])
plt.title('Jumlah Carports')
plt.xlabel('Carports')
plt.show()

# Boxplot for 'Price'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price_in_rp'])
plt.title('Price (in Rp)')
plt.xlabel('Price')
plt.show()
```

## Data Binning

```python
# Binning price diilakukan agar distribusi yang sangat lebar lebih mudah dianalisis, interpretasi jadi sederhana (Low/Medium/High)
bins_price = [0, 3_000_000_000, 5_000_000_000, df['price_in_rp'].max()]
labels_price = ['Low', 'Medium', 'High']

df['price_category'] = pd.cut(df['price_in_rp'], bins=bins_price, labels=labels_price, include_lowest=True)

# melihat hasil binning
df.head()
```

```python
# Bar plot Price Category vs Number of Houses

plt.figure(figsize=(6,4))
sns.countplot(x='price_category', data=df, order=['Low','Medium','High'])
plt.title('Distribution of Houses by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Number of Houses')
plt.show()

print('Bar plot tersebut menunjukkan distribusi rumah mayoritas di kategori Low, paling sedikit di Medium, dan High lebih tinggi sedikit dari Medium, \nyang menunjukkan pasar perumahan di Jabodetabek didominasi oleh rumah dengan harga relatif rendah')
```

```python
# Binning harga per meter persegi dilakukan untuk melihat apakah properti yang harga totalnya tinggi tapi harga per m²-nya relatif murah, 
# sehingga memberikan perspektif lebih detail dibanding hanya melihat harga keseluruhan.

df['price_per_m2'] = df['price_in_rp'] / df['land_size_m2'].replace(0, 1)
bins_ppm = [0, 10_000_000, 30_000_000, df['price_per_m2'].max()]
labels_ppm = ['Affordable', 'Mid-range', 'Premium']

df['price_per_m2_category'] = pd.cut(
    df['price_per_m2'],
    bins=bins_ppm,
    labels=labels_ppm,
    include_lowest=True
)

df.head()
```

```python
# Bar plot Price per m2 Category vs Number of Houses

plt.figure(figsize=(6,4))
sns.countplot(x='price_per_m2_category', data=df, order=['Affordable','Mid-range','Premium'])
plt.title('Distribution of Houses by Price per m² Category')
plt.xlabel('Price per m² Category')
plt.ylabel('Number of Houses')
plt.show()

print('Sebagian besar rumah berada di kategori Mid-range, meskipun banyak harga total yang tergolong Low. \nRumah dengan harga per m² Affordable jumlahnya cukup banyak, sedangkan yang Premium relatif sedikit, \nmenandakan bahwa properti dengan harga per m² sangat tinggi lebih jarang ditemukan di pasar Jabodetabek.')
```

```python
# Binning building age dilakukan agar lebih mudah dianalisis, divisualisasi, dan dibandingkan antar kelompok (New/Moderate/Old)

bins_age = [0, 2015, 2020, df['year_built'].max()]
labels_age = ['New', 'Moderate', 'Old']

df['year_built_category'] = pd.cut(
    df['year_built'],
    bins=bins_age,
    labels=labels_age,
    include_lowest=True
)

df.head()
```

```python
# Bar plot Building Age Category vs Number of Houses

plt.figure(figsize=(6,4))
sns.countplot(x='year_built_category', data=df, order=['New','Moderate','Old'])
plt.title('Distribution of Houses by Year Built Category')
plt.xlabel('Year Built Category')
plt.ylabel('Number of Houses')
plt.show()

print('Mayoritas rumah tergolong Old, diikuti dengan New, dan yang paling sedikit Moderate. \nHal ini menunjukkan sebagian besar properti di daerah Jabodetabek sebagian besar berumur lebih dari 10 tahun sekarang.' \
' \nNamun, data ini kurang akurat karena banyak nilai building_age yang diisi dengan median.')
```

```python
# Binning land size dilakukan agar distribusi yang sangat lebar lebih mudah dianalisis, interpretasi jadi sederhana (Small/Medium/Large)

bins_land = [0, 100, 300, df['land_size_m2'].max()]
labels_land = ['Small', 'Medium', 'Large']

df['land_size_category'] = pd.cut(
    df['land_size_m2'],
    bins=bins_land,
    labels=labels_land,
    include_lowest=True
)

df.head()
```

```python
# Bar plot Land Size Category vs Number of Houses

plt.figure(figsize=(6,4))
sns.countplot(x='land_size_category', data=df, order=['Small','Medium','Large'])
plt.title('Distribution of Houses by Land Size')
plt.xlabel('Land Size Category')
plt.ylabel('Number of Houses')
plt.show()

print('Mayoritas rumah berada di kategori Small, diikuti Medium, dan Large paling sedikit. \nHal ini menunjukkan bahwa sebagian besar properti di pasar Jabodetabek memiliki ukuran lahan yang relatif kecil hingga sedang.')
```

```python
# Binning land size dilakukan agar distribusi yang sangat lebar lebih mudah dianalisis, interpretasi jadi sederhana (Small/Medium/Large)

bins_land = [0, 100, 300, df['building_size_m2'].max()]
labels_land = ['Small', 'Medium', 'Large']

df['building_size_category'] = pd.cut(
    df['building_size_m2'],
    bins=bins_land,
    labels=labels_land,
    include_lowest=True
)

df.head()
```

```python
# Bar plot Building Size Category vs Number of Houses

plt.figure(figsize=(6,4))
sns.countplot(x='building_size_category', data=df, order=['Small','Medium','Large'])
plt.title('Distribution of Houses by Building Size')
plt.xlabel('Building Size Category')
plt.ylabel('Number of Houses')
plt.show()

print('Mayoritas rumah berada di kategori Small, diikuti Medium, dan Large paling sedikit. \nHal ini menunjukkan bahwa sebagian besar properti di pasar Jabodetabek memiliki ukuran bangunan yang relatif kecil hingga sedang.')
```

```python
# Correlation heatmap untuk membantu mengidentifikasi variabel numerik yang paling berpengaruh terhadap harga properti, 
# sehingga dapat menjadi dasar dalam analisis penentuan nilai dan strategi pemasaran.

plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=['float64', 'int32']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

#### Key Findings/Insights
##### Berdasarkan Correlation Heatmap di atas, terlihat bahwa variabel yang paling kuat berhubungan dengan harga properti (price_in_rp) adalah luas bangunan (building_size_m2, r=0.80) dan luas tanah (land_size_m2, r=0.73), yang menunjukkan bahwa ukuran fisik properti menjadi faktor utama. Harga per meter persegi (price_per_m2, r=0.49) juga berkorelasi cukup kuat, meskipun tidak sekuat luas bangunan atau tanah. Jumlah kamar mandi (bathrooms, r=0.56) dan kamar tidur (bedrooms, r=0.50) memiliki hubungan positif dengan harga, yang mendukung argumen bahwa semakin besar rumah dengan fasilitas lebih banyak, semakin tinggi juga harganya. Korelasi antarvariabel juga terlihat jelas, seperti antara kamar tidur dan kamar mandi (r=0.77), serta antara maid_bedrooms dan maid_bathrooms (r=0.74), yang menggambarkan konsistensi dalam desain rumah. Sementara itu, variabel seperti tahun dibangun (year_built, r=-0.21) dan koordinat geografis (lat/long, r≈0.20 atau lebih rendah) memiliki korelasi yang sangat lemah terhadap harga, menandakan bahwa faktor fisik properti lebih dominan dibanding usia bangunan atau lokasi geografis secara umum.
